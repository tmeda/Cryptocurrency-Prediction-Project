```python
import glob
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv(f'./../../data/train/arima_model.csv')
df=df.set_index('date')
```

```python
#creating the train and validation set
train, test = train_test_split(df,test_size = 0.011,shuffle = False)

#fit the model
from statsmodels.tsa.vector_ar.var_model import VAR
from statsmodels.tsa.vector_ar.var_model import VARResults

model = VAR(endog=train.values)
model_fit = model.fit( ic='aic')

# make prediction on validation
prediction = model_fit.forecast(model_fit.y, steps=len(test))
preds=pd.DataFrame(prediction,columns=test.columns,index=test.index).shift(-1)[:-1]
```

```python
model_fit.summary()
```

```python
targets=['bit_close','eth_close','ada_close','bnb_close']
preds=preds[targets]
```

```python
temp=[]
for i in preds.columns:
    temp.append(i+'_multivariate_arima')
preds.columns=temp
```

```python
preds.to_csv(f'../../data/predictions/multivariate_arima.csv')
```
